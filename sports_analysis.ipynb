```python
player = df[df["Player"] == "Andrew Shaw"]
print(player[player.columns[5:21]])
playerData = player[player.columns[5:21]]
list_columns = list(df.columns[5:21])
for index, c in enumerate(list_columns):
    if 'ATOI' in c:
        list_columns.pop(index)
x = df[list_columns]
y = df['Player']
model = runKNN(df,x,y,1)
relatedPlayers = model.kneighbors(playerData, n_neighbors=3, return_distance=False)
for p in relatedPlayers[0]: 
    print(df.iloc[int(p)]["Player"])
    print(df.iloc[int(p)])
```

All the similar players to Andrew Shaw seems to share the similar amount of points and plus/minus which means each of these players have similar impact on the team as Andrew Shaw

```python
player = df[df["Player"] == "Brett Connolly"]
print(player[player.columns[5:21]])
playerData = player[player.columns[5:21]]
list_columns = list(df.columns[5:21])
for index, c in enumerate(list_columns):
    if 'ATOI' in c:
        list_columns.pop(index)
x = df[list_columns]
y = df['Player']
model = runKNN(df,x,y,1)
relatedPlayers = model.kneighbors(playerData, n_neighbors=3, return_distance=False)
for p in relatedPlayers[0]: 
    print(df.iloc[int(p)]["Player"])
    print(df.iloc[int(p)])
```
