```python
from utils import * 
import pandas as pd
import altair as alt
import chardet
```

```python

with open(r'nhl_stats.csv', 'rb') as f:
    result = chardet.detect(f.read())
    df = pd.read_csv(r'nhl_stats.csv', encoding = result['encoding'])
new_header = df.iloc[0]
df = df[1:]
df.columns = new_header
df.head(20)
```

```python
split_data = df['Player'].str.split('\\',1)
```

```python
all_data = split_data.to_list()
real_names = []
for i in all_data:
    real_names.append(i[0])
df['Player'] = real_names
```

```python
print(type(df['Player'].iloc[0]))
df['PTS'] = df['PTS'].apply(pd.to_numeric)
print(type(df['PTS'].iloc[0]))
```

```python
alt.Chart(df.reset_index()).mark_bar().encode(
    x = 'Player',
    y = 'PTS'
   )
```
